## Análise e Resultados

```python
#importações
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
import os
```

```python
#carregando a base de dados
dados_2009 = pd.read_table("DOM2009.csv", encoding="windows-1252", sep=",")

dados_2013 = pd.read_table("DOM2013.csv", encoding="windows-1252", sep=",")
```

```python
#criando as categorias
dados_2009.V0201 = dados_2009.V0201.astype('category')
dados_2009.V4105 = dados_2009.V4105.astype('category')
dados_2009.V2103 = dados_2009.V2103.astype('category')
dados_2009.V2105 = dados_2009.V2105.astype('category')
dados_2009.V2107 = dados_2009.V2107.astype('category')
dados_2009.V2109 = dados_2009.V2109.astype('category')

dados_2013.V0201 = dados_2013.V0201.astype('category')
dados_2013.V4105 = dados_2013.V4105.astype('category')
dados_2013.V2103 = dados_2013.V2103.astype('category')
dados_2013.V2105 = dados_2013.V2105.astype('category')
dados_2013.V2107 = dados_2013.V2107.astype('category')
dados_2013.V2109 = dados_2013.V2109.astype('category')
```

```python
#Conversão Legenda - Descrição 
dados_2009.V0201.cat.categories = ('Particular Permanente','Particular Improvisado','Coletivo')
dados_2009.V4105.cat.categories = ('URBANA - Cidade ou vila, área urbanizada','URBANA - Cidade ou vila, área não urbanizada','URBANA - Área urbana isolada','RURAL - Aglomerado rural de extensão urbana','RURAL - Aglomerado rural, isolado, povoado','RURAL - Aglomerado rural, isolado, núcleo','RURAL - Aglomerado rural, isolado, outros aglomerados','RURAL - Zona rural exclusive aglomerado rural')
dados_2009.V2103.cat.categories = ('Sim','Não')
dados_2009.V2105.cat.categories = ('Sim','Não')
dados_2009.V2107.cat.categories = ('Sim','Não')
dados_2009.V2109.cat.categories = ('Sim','Não')

dados_2013.V0201.cat.categories = ('Particular Permanente','Particular Improvisado','Coletivo')
dados_2013.V4105.cat.categories = ('URBANA - Cidade ou vila, área urbanizada','URBANA - Cidade ou vila, área não urbanizada','URBANA - Área urbana isolada','RURAL - Aglomerado rural de extensão urbana','RURAL - Aglomerado rural, isolado, povoado','RURAL - Aglomerado rural, isolado, núcleo','RURAL - Aglomerado rural, isolado, outros aglomerados','RURAL - Zona rural exclusive aglomerado rural')
dados_2013.V2103.cat.categories = ('Sim','Não')
dados_2013.V2105.cat.categories = ('Sim','Não')
dados_2013.V2107.cat.categories = ('Sim','Não')
dados_2013.V2109.cat.categories = ('Sim','Não')
```

```python
#1° recorte: estado de são paulo
tabela_2009 = dados_2009[dados_2009.UF>20]

tabela_2013 = dados_2013[dados_2013.UF>20]
```

```python
#corte da base de dados, mantendo apenas as colunas desejadas
tabela_cortada_2009 = tabela_2009.iloc[:,[6,8,44,58,70,71,72,73]]

tabela_cortada_2013 = tabela_2013.iloc[:,[6,8,59,72,81,82,83,84]]
```

```python
#retirando valores indefinidos das variáveis de interesse
tabela_renda_2009 = tabela_cortada_2009[(tabela_cortada_2009.V4614 != 999999999999)]
tabela_renda_2009.V0105.dropna()
tabela_renda_2009.V0201.dropna()
tabela_renda_2009.V4105.dropna()
tabela_renda_2009.V2103.dropna()
tabela_renda_2009.V2105.dropna()
tabela_renda_2009.V2107.dropna()
tabela_renda_2009.V2109.dropna()
tabela_renda_2013 = tabela_cortada_2013[(tabela_cortada_2013.V4614 != 999999999999)]
tabela_renda_2013.V0105.dropna()
tabela_renda_2013.V0201.dropna()
tabela_renda_2013.V4105.dropna()
tabela_renda_2013.V2103.dropna()
tabela_renda_2013.V2105.dropna()
tabela_renda_2013.V2107.dropna()
tabela_renda_2013.V2109.dropna()
```

```python
#analisando numericamente

#Moradores
moradores_2009 = (tabela_renda_2009.V0105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(moradores_2009)


moradores_2013 = (tabela_renda_2013.V0105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(moradores_2013)


#arranjo em ordem crescente do número de moradores - plotagem do gráfico de barras
moradores_2009.sort_index(inplace=True)
moradores_2013.sort_index(inplace=True)
```

```python
#Quanto a moradia
conta_por_casa_2009 = (tabela_renda_2009.V0201.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(conta_por_casa_2009)

conta_por_casa_2013 = (tabela_renda_2013.V0201.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(conta_por_casa_2013)
```

```python
#Quanto a zona
conta_por_zona2009 = (tabela_renda_2009.V4105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(conta_por_zona2009)

conta_por_zona2013 = (tabela_renda_2013.V4105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(conta_por_zona2013)
```

```python
#Os moradores tiveram a preocupação de que os alimentos acabassem antes de poderem comprar ou receber mais comida
preocupacao_alimenta_2009 = (tabela_renda_2009.V2103.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(preocupacao_alimenta_2009)

preocupacao_alimenta_2013 = (tabela_renda_2013.V2103.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(preocupacao_alimenta_2013)

```

```python
#Os alimentos acabaram antes que os moradores deste domicílio tivessem dinheiro para comprar mais comida
sem_dinheiro_nem_alimento_2009 = (tabela_renda_2009.V2105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(sem_dinheiro_nem_alimento_2009)

sem_dinheiro_nem_alimento_2013 = (tabela_renda_2013.V2105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(sem_dinheiro_nem_alimento_2013)
```

```python
#Os moradores ficaram sem dinheiro para ter uma alimentação saudável e variada
sem_dinheiro_para_alimentacao_saudavel_2009 = (tabela_renda_2009.V2107.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(sem_dinheiro_para_alimentacao_saudavel_2009)

sem_dinheiro_para_alimentacao_saudavel_2013 = (tabela_renda_2013.V2107.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(sem_dinheiro_para_alimentacao_saudavel_2013)
```

```python
#Os moradores deste domicílio comeram apenas alguns alimentos que ainda tinham porque o dinheiro acabou
sem_dinheiro_para_comida_2009 = (tabela_renda_2009.V2109.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(sem_dinheiro_para_comida_2009)

sem_dinheiro_para_comida_2013 = (tabela_renda_2013.V2109.value_counts(sort=False, normalize=True)*100).round(decimals=2)
print(sem_dinheiro_para_comida_2013)
```

```python
# Gráficos
plot = moradores_2009.plot(kind='bar',title='Numero de Moradores 2009',figsize=(6, 6))
plt.ylabel('Freq.Relativa (em %)')
plt.show()

plot = moradores_2013.plot(kind='bar',title='Numero de Moradores 2013',figsize=(6, 6))
plt.ylabel('Freq.Relativa (em %)')
plt.show()
```

```python
plot = preocupacao_alimenta_2009.plot(kind='pie',title='Preocupação alimentar 2009',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

plot = preocupacao_alimenta_2013.plot(kind='pie',title='Preocupação alimentar 2013',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plot = sem_dinheiro_nem_alimento_2009.plot(kind='pie',title='Os alimentos acabaram 2009',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

plot = sem_dinheiro_nem_alimento_2013.plot(kind='pie',title='Os alimentos acabaram 2013',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plot = sem_dinheiro_para_alimentacao_saudavel_2009.plot(kind='pie',title='moradores ficaram sem dinheiro para uma alimentação saudável e variada 2009',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

plot = sem_dinheiro_para_alimentacao_saudavel_2013.plot(kind='pie',title='moradores ficaram sem dinheiro para uma alimentação saudável e variada 2013',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plot = sem_dinheiro_para_comida_2009.plot(kind='pie',title='comeram apenas alguns alimentos que ainda tinham, porque o dinheiro acabou 2009',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

plot = sem_dinheiro_para_comida_2013.plot(kind='pie',title='comeram apenas alguns alimentos que ainda tinham, porque o dinheiro acabou 2013',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
