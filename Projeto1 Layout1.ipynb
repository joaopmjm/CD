## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#Se usar o arquivo descompactado
#pd.read_csv('DOM2015.csv',sep=',')
planilha2009 = pd.read_csv('DOM2009csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
defe = pd.DataFrame()

defe['medo alimentar'] = planilha2009.V2103

defe['UF'] = planilha2009[planilha2009.UF>2].UF

defe['sem dinheiro'] = planilha2009.V2107

defe['deficiência alimentar'] = planilha2009.V2109

defe['sem alimento'] = planilha2009.V2105

defe['tipo residencial'] = planilha2009.V4105

defe_no = defe.dropna()
```

```python
#filtragem]
defe.loc[planilha2009.UF>10,'UF'] = 'Norte'
defe.loc[planilha2009.UF>20,'UF'] = 'Nordeste'
defe.loc[planilha2009.UF>30,'UF'] = 'Sudeste'
defe.loc[planilha2009.UF>40,'UF'] = 'Sul'
defe.loc[planilha2009.UF>50,'UF'] = 'Centro-Oeste'
defe.loc[planilha2009.UF==53,'UF'] = 'Distrito Ferderal'

defe.loc[planilha2009.V2103 == 1,'medo alimentar'] = 'sim'
defe.loc[planilha2009.V2103 == 3,'medo alimentar'] = 'não'

defe.loc[planilha2009.V2107 == 1,'sem dinheiro'] = 'sim'
defe.loc[planilha2009.V2107 == 3,'sem dinheiro'] = 'não'

defe.loc[planilha2009.V2109 == 1,'deficiência alimentar'] = 'sim'
defe.loc[planilha2009.V2109 == 3,'deficiência alimentar'] = 'não'

defe.loc[planilha2009.V2105 == 1,'sem alimento'] = 'sim'
defe.loc[planilha2009.V2105 == 3,'sem alimento'] = 'não'

defe.loc[planilha2009.V4105 == 1,'tipo residencial'] = 'sim'
defe.loc[planilha2009.V4105 == 3,'tipo residencial'] = 'não'

defe = defe.dropna()
```

```python
#Sem dinheiro para comida em diferentes regiões
defe['medo alimentar'].value_counts().plot(kind='pie',autopct="%0.2f", legend=True)
plt.show()
```

```python
defe['sem dinheiro'].value_counts().plot(kind='pie',autopct="%0.2f",title = 'sem dinheiro para alimentação', legend = True)
```

```python
plt.subplot(121)
defe['sem alimento'].value_counts().plot(kind='pie',autopct="%0.2f",title = 'sem alimento', legend = False)
plt.subplot(122)
defe['medo alimentar'].value_counts().plot(kind='pie',autopct="%0.2f",title = 'medo alimentar', legend=False)
```
